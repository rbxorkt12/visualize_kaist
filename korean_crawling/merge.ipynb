```python
import pandas as pd
import os
```

```python
path_dir = 'pickle files'
file_list = os.listdir(path_dir)
file_list
```

```python
df = pd.DataFrame(columns=['url', 'title', 'content'])
for file in file_list:
    if file == '.ipynb_checkpoints':
        continue
    file = 'pickle files/' + file
    temp = pd.read_pickle(file)
    print(len(temp))
    df = pd.concat([df, temp], ignore_index=True)
df
```

```python
df.to_excel('articles/merged.xlsx')
df.to_pickle('pickle files/merged.pkl')
```
